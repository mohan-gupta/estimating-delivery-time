```python
#importing libraries
import pandas as pd
import os

from tqdm.notebook import tqdm

from collections import defaultdict
```

```python
train_file_names = os.listdir("../dataset/train")
test_file_names = os.listdir("../dataset/test")
```

```python
train_file_paths = ["../dataset/train/"+name for name in train_file_names]
test_file_paths = ["../dataset/test/"+name for name in test_file_names]
```

```python
def read_data(file_path):
    """function to read the file and return its contents"""
    with open(file_path) as f:
        data = f.read()
    return data.split("\n")[:-1]
```

```python
#sample data
read_data(train_file_paths[0])
```

```python
def convert_to_df(file_paths):
    """function to convert file content to data frame"""
    data_dct = defaultdict(list)
    for file_path in tqdm(file_paths):
        data = read_data(file_path)
        for row in data:
            row_val = row.split(" ")
            data_dct[row_val[0]].append(row_val[-1])
            
    return pd.DataFrame(data_dct)
```

```python
train_df = convert_to_df(train_file_paths)
```

```python
test_df = convert_to_df(test_file_paths)
```
